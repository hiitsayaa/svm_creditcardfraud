**Cek Extreme Outlier Semua Variable**

```python
# pisahkan fitur dan target
X = df.drop(columns=["Class"])
y = df["Class"]

def detect_extreme_outliers_iqr(data, target):
    outlier_summary = []

    for col in data.columns:
        if data[col].dtype != 'object':

            # Hitung IQR
            Q1 = data[col].quantile(0.25)
            Q3 = data[col].quantile(0.75)
            IQR = Q3 - Q1

            # Tentukan batas extreme outlier
            lower_bound = Q1 - 3 * IQR
            upper_bound = Q3 + 3 * IQR

            # Deteksi data outlier
            outlier_mask = (data[col] < lower_bound) | (data[col] > upper_bound)
            outliers = data[outlier_mask]

            # Ambil label (fraud / non-fraud) dari outlier
            outlier_classes = target[outlier_mask]

            # Hitung jumlah outlier
            num_outliers = len(outliers)
            num_fraud = (outlier_classes == 1).sum()
            num_non_fraud = (outlier_classes == 0).sum()

            # Hasil
            outlier_summary.append({
                "feature": col,
                "num_outliers": num_outliers,
                "num_non_fraud": num_non_fraud,
                "num_fraud": num_fraud,
                "percentage": (num_outliers / len(data)) * 100
            })

    return pd.DataFrame(outlier_summary)

# jalankan
iqr_outliers = detect_extreme_outliers_iqr(X,y)
print(iqr_outliers)
```

# **Imbalance Handling**

**Cek Imbalance**

```python
from sklearn.model_selection import train_test_split

# Ambil subset 50k dari training set dengan stratify
X_train_small, _, y_train_small, _ = train_test_split(
    X_train, y_train,
    train_size=50000,
    stratify=y_train,
    random_state=42
)

# Cek distribusi kelas
print(y_train_small.value_counts())
print(y_train_small.value_counts(normalize=True) * 100)
```

```python
from sklearn.svm import SVC

model_imbalance = SVC(kernel='rbf', probability=True, random_state=42)
model_imbalance.fit(X_train_small, y_train_small)
```

```python
from sklearn import metrics

# --- Training evaluation ---
y_train_pred = model_imbalance.predict(X_train)
y_train_proba = model_imbalance.predict_proba(X_train)[:,1]

acc_train = metrics.accuracy_score(y_train, y_train_pred)
prec_train = metrics.precision_score(y_train, y_train_pred)
rec_train = metrics.recall_score(y_train, y_train_pred)
f1_train = metrics.f1_score(y_train, y_train_pred)
auc_train = metrics.roc_auc_score(y_train, y_train_proba)
cm_train = metrics.confusion_matrix(y_train, y_train_pred)

# --- Testing evaluation ---
y_test_pred = model_imbalance.predict(X_test)
y_test_proba = model_imbalance.predict_proba(X_test)[:,1]

acc_test = metrics.accuracy_score(y_test, y_test_pred)
prec_test = metrics.precision_score(y_test, y_test_pred)
rec_test = metrics.recall_score(y_test, y_test_pred)
f1_test = metrics.f1_score(y_test, y_test_pred)
auc_test = metrics.roc_auc_score(y_test, y_test_proba)
cm_test = metrics.confusion_matrix(y_test, y_test_pred)

# --- Gabungkan ke tabel ---
results = pd.DataFrame({
    "Accuracy":[acc_train, acc_test],
    "Precision":[prec_train, prec_test],
    "Recall":[rec_train, rec_test],
    "F1-score":[f1_train, f1_test],
    "AUC":[auc_train, auc_test]
}, index=["Training","Testing"])

print("=== Metrics Training vs Testing (Imbalanced) ===")
print(results)

# --- Confusion Matrix print biasa ---
print("\nConfusion Matrix - Training:\n", cm_train)
print("\nConfusion Matrix - Testing:\n", cm_test)

# --- ROC Curve gabungan ---
fpr_train, tpr_train, _ = metrics.roc_curve(y_train, y_train_proba)
fpr_test, tpr_test, _ = metrics.roc_curve(y_test, y_test_proba)

plt.figure(figsize=(8,6))
plt.plot(fpr_train, tpr_train, label=f"Train AUC = {auc_train:.5f}", color="blue")
plt.plot(fpr_test, tpr_test, label=f"Test AUC = {auc_test:.5f}", color="red")
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Imbalanced (Training vs Testing)")
plt.legend()
plt.show()
```

# Hyperparameter Tuning

**Hyperparameter Tuning dengan GridSearchCV (Model SMOTE)**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore') # untuk mengabaikan peringatan yang tidak penting

# Mempersiapkan subset dari data SMOTE untuk tuning yang lebih cepat
# Mengambil sekitar 10% dari data SMOTE untuk mempercepat proses tuning
X_smote_small, _, y_smote_small, _ = train_test_split(
    X_smote, y_smote, test_size=0.9, stratify=y_smote, random_state=42
)

print(f"Shape data SMOTE kecil untuk tuning: {X_smote_small.shape}")
print(f"Distribusi kelas di data SMOTE kecil:\n{y_smote_small.value_counts()}")

# Definisikan parameter grid untuk SVC
# Rentang parameter yang lebih kecil untuk demo, bisa diperluas untuk hasil optimal
param_grid = {
    'C': [0.1, 1, 10],
    'gamma': ['scale', 'auto'],
    'kernel': ['rbf']
}

# Inisialisasi model SVC
svc_model_for_tuning = SVC(probability=True, random_state=42)

# Inisialisasi GridSearchCV
# Scoring menggunakan 'f1' atau 'recall' karena ini adalah masalah imbalance
grid_search = GridSearchCV(
    estimator=svc_model_for_tuning,
    param_grid=param_grid,
    scoring='f1',  # Atau 'recall' tergantung prioritas Anda
    cv=3,          # Jumlah fold untuk cross-validation, 3-5 sudah cukup baik
    n_jobs=-1,     # Gunakan semua core CPU yang tersedia
    verbose=2      # Menampilkan progress
)

print("Memulai proses GridSearchCV...")
grid_search.fit(X_smote_small, y_smote_small)

print("\nProses GridSearchCV selesai.")

# Cetak parameter terbaik dan skor terbaik
print("Parameter terbaik ditemukan:", grid_search.best_params_)
print("Skor F1 terbaik (rata-rata CV):", grid_search.best_score_)

# Ambil model terbaik dari grid search
best_smote_model = grid_search.best_estimator_
```
